```python
## fetching the aggredated transaction data from imported github file
path_a = "C:/Users/rdine/Data_Science/data_science_practise/DataScience_projects/phone_pe/pulse/data/aggregated/transaction/country/india/state/"
agg_trans_file = os.listdir(path_a)

columns1 = {"States":[], "Years":[], "Quarter":[], "Transaction_type":[], "Transaction_count":[], "Transaction_amount":[] }

for state in agg_trans_file:
    current_states = path_a+state+"/"  ## accessing the individual states
    agg_year = os.listdir(current_states) ##accessing the states year

    for year in agg_year:
        current_year = current_states+year+"/"  ## accessing individual states year
        agg_file_lists = os.listdir(current_year) ## accessing the years json files

        for file in agg_file_lists:
            current_file = current_year+file
            data = open(current_file,"r")

            data_A = json.load(data)
            for i in data_A["data"]["transactionData"]:
                name = i["name"]
                count = i["paymentInstruments"][0]["count"]
                amount = i["paymentInstruments"][0]["amount"]
                columns1["Transaction_type"].append(name)
                columns1["Transaction_count"].append(count)
                columns1["Transaction_amount"].append(amount)
                columns1["States"].append(state)
                columns1["Years"].append(year)
                columns1["Quarter"].append(int(file.strip(".json")))
```

```python

aggregation_transaction = pd.DataFrame(columns1)
# aggre_transaction = aggregation_transaction
aggregation_transaction["States"] = aggregation_transaction["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
aggregation_transaction["States"] = aggregation_transaction["States"].str.replace("-"," ")
aggregation_transaction["States"] = aggregation_transaction["States"].str.title()
aggregation_transaction['States'] = aggregation_transaction['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")

aggregation_transaction

```
